```python
import pandas as pd
df = pd.read_csv('D:/KAtE/DATA-science/exam/fifa_players.csv', low_memory=False)
df.head()
```

```python
#1. Топ 100 гравців з найвищим рейтингом. (1 бал)
top_100_rated_players = df[['short_name', 'overall', 'club_name', 'wage_eur']].sort_values(by='overall', ascending=False).head(100)
print(top_100_rated_players)
```

```python
#2.Топ 100 гравців з найвищими зарплатами. 
top_100_highest_wage_players = df[['short_name', 'overall', 'club_name', 'wage_eur']].sort_values(by='wage_eur', ascending=False).head(100)
print(top_100_highest_wage_players)

#Порівняти з п.1 чи відповідають гравці з найвищими рейтингами найвищим зарплатам і навпаки. (1 бал)
top_rated_names = set(top_100_rated_players['short_name'])
top_wage_names = set(top_100_highest_wage_players['short_name'])
common_players = top_rated_names.intersection(top_wage_names)
print(f'Number of players in both top 100 rated and top 100 highest paid: {len(common_players)}')
print(common_players)
```

```python
#3. Топ 30 воротарів з найвищим рейтингом. (1 бал)
top_30_goalkeepers = df[df['player_positions'].str.contains('GK')][['short_name', 'overall', 'club_name']].sort_values(by='overall', ascending=False).head(30)
print(top_30_goalkeepers)
```

```python
#4. Топ 30 команд з гравцями найвищого рейтингу в середньому.(1 бал)
top_30_teams_by_avg_rating = df.groupby('club_name')['overall'].mean().sort_values(ascending=False).head(30)
print(top_30_teams_by_avg_rating)
```

```python
#5. Топ 30 команд в яких в середньому гравці мають найбільшу швидкість. (1 бал)
top_30_teams_by_avg_pace = df.groupby('club_name')['pace'].mean().sort_values(ascending=False).head(30)
print(top_30_teams_by_avg_pace)
```

```python
#6. Топ усіх Ліг в яких зібрані найкращі дриблери. (dribbling) (2 бали)
top_leagues_by_dribbling = df.groupby('league_name')['dribbling'].mean().sort_values(ascending=False)
print(top_leagues_by_dribbling)
```

```python
#7. Топ 30 команд з гравцями найвищого рейтингу враховуючи, що команда складається з 11 гравців, серед яких обовʼязково 1 воротар, 4 захисники, 4 півзахисники, 2 нападники. (3 бали)
#гравці по позиціях
gk = df[df['player_positions'].str.contains('GK')]  #воротарі
df_defenders = df[df['player_positions'].str.contains('CB|LB|RB|LWB|RWB')]  #захисники
mf_midfielders = df[df['player_positions'].str.contains('CM|CDM|CAM|LM|RM')]  #півзахисники
fw_forwards = df[df['player_positions'].str.contains('ST|LW|RW|CF')]  #нападники

#фільтрую команди, які мають обов'язково 1 воротаря, 4 захисників, 4 півзахисників і 2 нападників
valid_teams = set.intersection(
    set(gk['club_name'].value_counts()[gk['club_name'].value_counts() >= 1].index),
    set(df_defenders['club_name'].value_counts()[df_defenders['club_name'].value_counts() >= 4].index),
    set(mf_midfielders['club_name'].value_counts()[mf_midfielders['club_name'].value_counts() >= 4].index),
    set(fw_forwards['club_name'].value_counts()[fw_forwards['club_name'].value_counts() >= 2].index)
)
#вибираю гравців з цих команд
gk_filtered = gk[gk['club_name'].isin(valid_teams)]
df_filtered = df_defenders[df_defenders['club_name'].isin(valid_teams)]
mf_filtered = mf_midfielders[mf_midfielders['club_name'].isin(valid_teams)]
fw_filtered = fw_forwards[fw_forwards['club_name'].isin(valid_teams)]

#топ гравців для кожної команди
gk_top = gk_filtered.sort_values(by='overall', ascending=False).groupby('club_name').head(1).reset_index(drop=True)
df_top = df_filtered.sort_values(by='overall', ascending=False).groupby('club_name').head(4).reset_index(drop=True)
mf_top = mf_filtered.sort_values(by='overall', ascending=False).groupby('club_name').head(4).reset_index(drop=True)
fw_top = fw_filtered.sort_values(by='overall', ascending=False).groupby('club_name').head(2).reset_index(drop=True)

#об'єдную всіх гравців по кожній команді
top_teams_players = pd.concat([gk_top, df_top, mf_top, fw_top]).reset_index(drop=True)

#виводжу середній рейтинг для кожної команди
team_avg_rating = top_teams_players.groupby('club_name')['overall'].mean().sort_values(ascending=False).head(30)

print(team_avg_rating)
```

```python
#11. Придумати метрику, яка оцінюватиме якість пенальтиста. (2 бали)
#Я вважаю, що для оцінки якості пенальтиста важливими є такі навички: 
# 1. shooting: вміння стріляти по воротах, як допоміжна метрика - 0.10
# 2. attacking_finishing: Точність удару, що безпосередньо впливає на результативніст - 0.2
# 3. mentality_penalties: Специфічна навичка для виконання пенальті - 0.4, думаю, якщо вона "спеціфічна", то напевно дуже важлива
# 4. mentality_composure: Самовладання, яке важливе під час виконання пенальті, оскільки стрес може вплинути на результат - 0.3 - друга по значимості, мені здається
# Формула може виглядати так:
# Quality penalty=0.10×Shooting+0.20×Attacking Finishing+0.40×Mentality Penalties+0.30×Mentality Composure
```

```python
# Фільтруємо тільки пенальтистів
penalty_players_df = df[df['player_positions'].str.contains('ST|CF|RW|LW|CAM', na=False)]

penalty_players_df.loc[:,'penalty_quality'] = (
    0.10 * df['shooting'] +
    0.20 * df['attacking_finishing'] +
    0.40 * df['mentality_penalties'] +
    0.30 * df['mentality_composure']
)

print(penalty_players_df[['short_name', 'penalty_quality']].head())
#топ 10 пенальтистів
top_penalty_players = penalty_players_df[['short_name', 'penalty_quality']].sort_values(by='penalty_quality', ascending=False).head(10)
print(top_penalty_players)
```

```python
#12. Придумати метрику, яка оцінюватиме якість воротаря, який відбиває пенальті. (2 бали)
# 1. goalkeeping_diving: Вміння стрибати для відбиття ударів - 0.2
# 2. goalkeeping_handling: Здатність ловити м'яч -0.2
# 3. goalkeeping_positioning: Позиціювання на полі -0.2
# 4. goalkeeping_reflexes: Реакція воротаря, на мою думку йде в першу чергу - 0.4

# Формула може виглядати так:
# Quality goalkeeper=0.20×goalkeeping_diving+0.20×goalkeeping_handling+0.20×goalkeeping_positioning+0.40×goalkeeping_reflexes
```

```python
# Фільтруємо тільки воротарів
goalkeepers_df = df[df['player_positions'].str.contains('GK')]

# Додаємо стовпчик якості воротарів
goalkeepers_df.loc[:, 'goalkeeper_quality'] = (
    0.20 * goalkeepers_df['goalkeeping_diving'] +
    0.20 * goalkeepers_df['goalkeeping_handling'] +
    0.20 * goalkeepers_df['goalkeeping_positioning'] +
    0.40 * goalkeepers_df['goalkeeping_reflexes']
)

print(goalkeepers_df[['short_name', 'goalkeeper_quality']].head())
#топ 10 воротарів
top_goalkeepers_players = goalkeepers_df[['short_name', 'goalkeeper_quality']].sort_values(by='goalkeeper_quality', ascending=False).head(10)
print(top_goalkeepers_players)
```

```python
#13. Чи є клуби, в яких пенальтисти суттєво (придумати, що значить “суттєво”) кращі за воротарів, які відбивають? А навпаки? (1 бал)

#“суттєво” можна визначити як різницю в середніх значеннях метрик для пенальтистів і воротарів, наприклад, більше ніж 0.5 або 1 бал
```

```python
# Середні значення якості для кожного клубу серед пенальтистів і воротарів (данні з попередніх завдань)
penalty_quality_by_club = penalty_players_df.groupby('club_name')['penalty_quality'].mean().reset_index()
goalkeeper_quality_by_club = goalkeepers_df.groupby('club_name')['goalkeeper_quality'].mean().reset_index()

# Об'єднання таблиць для порівняння
club_quality = pd.merge(penalty_quality_by_club, goalkeeper_quality_by_club, on='club_name', how='inner')

# Обчислення різниці
club_quality['difference'] = club_quality['penalty_quality'] - club_quality['goalkeeper_quality']

# Визначення суттєвих клубів
significant_better_penalty = club_quality[club_quality['difference'] > 0.5]
significant_better_goalkeeper = club_quality[club_quality['difference'] < -0.5]

#щоб бачити весь результат
pd.set_option('display.max_columns', None)
pd.set_option('display.expand_frame_repr', False)

print("Клуби, де пенальтисти суттєво кращі за воротарів:")
print(significant_better_penalty[['club_name', 'penalty_quality', 'goalkeeper_quality', 'difference']])

print("\nКлуби, де воротарі суттєво кращі за пенальтистів:")
print(significant_better_goalkeeper[['club_name', 'penalty_quality', 'goalkeeper_quality', 'difference']])

```

```python
#17. Зробити EDA з точки зору Дата Сайентиста. Які проблеми в датасеті? (У польових гравців можуть бути атрибути воротарів і навпаки)#
```

Проблеми датасету:
- Відсутні значення в атрибутах навичок, контрактах, національних позиціях та ін.
- Атрибути воротарів та польових гравців часто перетинаються, що може призводити до некоректного аналізу, якщо ці атрибути не будуть розділені або оброблені окремо.

Можна застосувати метод на основі кореляційної матриці. Ті змінні, які мають суттєву кореляцію з цільовою змінною (наприклад, рейтингом гравця), можна залишити. У нашому випадку порогове значення кореляції встановлено на рівні 0.3. Це дозволяє залишити змінні з помірною або високою кореляцією з цільовою змінною, що є доцільним для побудови моделі. Для цього:

1. Видаляємо змінні, які мають слабку кореляцію з цільовою ознакою (менше 0.3).
2. Якщо дві змінні мають дуже високу кореляцію між собою (наприклад, більше 0.9), можливо, варто залишити лише одну з них, щоб уникнути надлишковості даних.

```python
# Видаляємо нечислові колонки
numeric_df = df.select_dtypes(include=[np.number])

# Створюємо кореляційну матрицю для числових ознак
correlation_matrix = numeric_df.corr()

# Відберемо тільки ті ознаки, які мають найбільший вплив на 'overall' (> 0.5)
important_corr = correlation_matrix.loc[(correlation_matrix['overall'] > 0.5) | (correlation_matrix['overall'] < -0.5), 
                                        (correlation_matrix['overall'] > 0.5) | (correlation_matrix['overall'] < -0.5)]

# Візуалізуємо фільтровану кореляційну матрицю
plt.figure(figsize=(10, 8))
sns.heatmap(important_corr, annot=True, cmap='coolwarm')
plt.title("Кореляційна матриця для найбільш значущих ознак")
plt.show()

# Кореляція відносно цільової змінної 'overall'
correlation_with_overall = correlation_matrix['overall'].sort_values(ascending=False)
print(correlation_with_overall)

# Відберемо ознаки, які мають кореляцію з 'overall' більше 0.3 (порогове значення можна змінювати)
selected_features = correlation_with_overall[correlation_with_overall > 0.3].index.tolist()
print("Відібрані ознаки для моделі:", selected_features)

# Створюємо новий датафрейм тільки з відібраними ознаками
selected_df = numeric_df[selected_features]
print(selected_df.head())
```

Я використала кореляційну матрицю, щоб визначити найбільш релевантні змінні. На основі кореляції з цільовою змінною (overall), були відібрані змінні з кореляцією більше ніж 0.3.

- Кореляційна матриця була відфільтрована, щоб показати лише найбільш значущі змінні.
- Змінні з високою кореляцією з цільовою змінною включають: потенціал, логарифм вартості, логарифм зарплати, пасування, реакцію, атакувальні та захисні якості.

Для цього завдання я обрала алгоритм класифікації Random Forest, оскільки він добре підходить для таких типів завдань, як передбачення категорій на основі числових та категоріальних ознак. Модель була навчена на ознаках, що описують гравців, і передбачала їх основну позицію на полі.

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import LabelEncoder
import numpy as np

#в попередніх завдання були видалення колонок, тому завантажую дата сет заново
df = pd.read_csv('D:/KAtE/DATA-science/exam/fifa_players.csv', low_memory=False)

#створюю комбіновані показники для атаки та захисту
df['attack_quality'] = df[['shooting', 'dribbling', 'passing']].mean(axis=1)
df['defense_quality'] = df[['defending', 'physic', 'pace']].mean(axis=1)

# Логарифм вартості та зарплати
df['log_value_eur'] = np.log(df['value_eur'] + 1)
df['log_wage_eur'] = np.log(df['wage_eur'] + 1)

# Використовуємо LabelEncoder для категоріальних змінних
label_encoder = LabelEncoder()

# Кодуємо текстові дані в числові
df['club_position_encoded'] = label_encoder.fit_transform(df['club_position'].fillna('None'))

# створюємо колонку 'primary_position' на основі колонки 'player_positions'
df['primary_position'] = df['player_positions'].apply(lambda x: x.split(',')[0].strip() if pd.notna(x) else 'None')

# Вибираємо ознаки для моделювання (включаємо club_position_encoded)
features_with_positions = df[selected_features + ['club_position_encoded']].fillna(0)  # Заповнюємо пропущені значення нулями
target = df['primary_position']  # Цільова змінна

# Поділ на тренувальні та тестові дані
X_train, X_test, y_train, y_test = train_test_split(features_with_positions, target, test_size=0.3, random_state=42)

# Побудова моделі Random Forest
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

# Передбачення на тестових даних
y_pred = rf_model.predict(X_test)

# Оцінка точності
print("Accuracy з використанням club_position_encoded:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```
